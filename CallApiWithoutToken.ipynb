# Call API on Github Platform

```python
import requests
import pandas as pd
```

## Initialize Git url 

```python
url = "https://api.github.com/search/repositories"
url += "?q=language:python"
# This url return all repositories which using python language
```

## Send GET query for get data

#### Sometimes you need to send a firts query for get Token before to get data or response 
#### With this url token is note required

```python
headers = {"Accept": "application/vnd.github.v3+json"} #specify which type of response body we wait 
res = requests.get(url, headers=headers) # send get query 
print(f"Status code: {res.status_code}") # query status 200 ==> good query 
```

## Save my response in variable and convert to dictionary

```python
json_response = res.json()
```

## Print key

```python
json_response.keys()
```

### Print information about each repository

```python
for responses in responses_dict :
    print(f"Repository id: {responses['id']}")
    print(f"Repository name: {responses['name']}")
    print(f"Owner: {responses['owner']['login']}")
    print(f"Created repository on: {responses['created_at']}")
    print(f"Updated repository on: {responses['updated_at']}")
    print(f"Github repository url: {responses['html_url']}")
    print(f"Repository description: {responses['description']}")
    print(f"Stars: {responses['stargazers_count']}")
    print()
    pass
```

## Create empty list 

```python
Repository_ID ,Repository_name, Owner , created_at , updated_at ,html_url ,description,stargazers_count = ([] for _ in range(8))
```

```python
for responses in responses_dict :
    Repository_ID.append(responses['id'])
    Repository_name.append(responses['name'])
    Owner.append(responses['owner'])
    created_at.append(responses['created_at'])
    updated_at.append(responses['updated_at'])
    html_url.append(responses['html_url'])
    description.append(responses['description'])
    stargazers_count.append(responses['stargazers_count'])
```

## Create DataFrame

```python
DataFrame = pd.DataFrame(list(zip(Repository_ID ,Repository_name, Owner , created_at , updated_at ,html_url ,description,stargazers_count)),\
                         columns = ['Repository_ID' ,'Repository_nam','Owner' , 'created_at' , 'updated_at' ,'html_url' ,'description','stargazers_count'])
# Zip for having mapped between list and columns 
```
